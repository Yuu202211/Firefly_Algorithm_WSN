```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Beam properties and loading
E = 200e9  # Young's modulus (Pa)
I = 1e-4  # Moment of inertia (m^4)
L = 2.0  # Length of the beam (m)
w = 500  # Uniform distributed load (N/m)

# Function to return derivatives
def beam_equations(x, y):
    y1, y2, y3, y4 = y
    dydx = [y2,
            y3,
            y4,
            -w / (E * I)]  # Change sign for correct deflection direction
    return dydx

# Initial conditions
y0 = [0, 0, 0, 0]  # For a cantilever, starting conditions usually represent zero deflection and slope at the base

# Spatial domain
x_span = [0, L]
x_eval = np.linspace(0, L, 100)  # Points at which to evaluate the solution

# Solve the system of differential equations
sol = solve_ivp(beam_equations, x_span, y0, t_eval=x_eval, method='RK45')

# Plot the results
plt.plot(sol.t, sol.y[0], label='Deflection (y)')
plt.plot(sol.t, sol.y[1], label='Slope (dy/dx)')
plt.title('Deflection and Slope of a Cantilever Beam')
plt.xlabel('Position along beam (m)')
plt.ylabel('Deflection (m) / Slope (rad)')
plt.legend()
plt.grid(True)
plt.show()
```

### Euler-Bernoulli Beam Equation

The Euler-Bernoulli equation for beam bending under a distributed load `w(x)` is given by:

$$ EI \frac{d^4y}{dx^4} = w(x) $$

Where:
- `E` is the Young's modulus of the material,
- `I` is the moment of inertia of the beam's cross-section about the neutral axis,
- `y(x)` is the deflection of the beam at position `x`,
- `w(x)` is the distributed load per unit length as a function of `x`.

For a uniform distributed load `w` and a point load `P` at the free end, `w(x)` can be described as a constant `w` along the length of the beam and an additional term representing the point load at `x = L`.

### System of Differential Equations

While the Euler-Bernoulli beam equation is a fourth-order differential equation, not a system of multiple first-order differential equations, in practical numerical analysis and simulation, higher-order differential equations are often converted into a system of first-order differential equations because many numerical methods (like the Runge-Kutta method) are designed to solve first-order systems.

To convert the fourth-order Euler-Bernoulli equation into a system of first-order equations, introduce variables for each derivative of `y` up to the third order:

1. Let $y_1 = y$ (beam deflection),
2. Let $y_2 = \frac{dy}{dx}$ (slope of the beam),
3. Let $y_3 = \frac{d^2y}{dx^2}$ (curvature of the beam, related to the bending moment),
4. Let $y_4 = \frac{d^3y}{dx^3}$ (related to the shear force).

The Euler-Bernoulli beam equation can then be represented as a system of first-order differential equations:

1. $$\frac{dy_1}{dx} = y_2$$,
2. $$\frac{dy_2}{dx} = y_3$$,
3. $$\frac{dy_3}{dx} = y_4$$,
4. $$\frac{dy_4}{dx} = \frac{w(x)}{EI}$$.

### Boundary Conditions

For a cantilever beam:
- At `x = 0` (fixed end), the deflection `y_1 = 0` and the slope `y_2 = 0`.
- The conditions at the free end (`x = L`) depend on the loading and require expressing the bending moment and shear force in terms of the derivatives of `y`.

### Solving the System

This system can be solved using numerical methods given the specific loading conditions, material properties (`E`), and geometric properties (`I`). The solution will provide the deflection profile `y(x)`, slope, bending moment, and shear force along the length of the beam.

The impact of the point load `P` at the free end can be incorporated into the boundary conditions or as a discontinuity in `w(x)` at `x = L`, depending on the formulation.

Solving this system numerically requires discretizing the beam into segments or using a finite element method (FEM) solver, which is common in engineering analysis software for complex loading and geometry scenarios.
